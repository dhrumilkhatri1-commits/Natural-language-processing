```python
# ==========================================
# NLP (Cosine Similarity)
# ==========================================

# Install required library
!pip install -q scikit-learn

# Import libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample text
text1 = "Python is a powerful programming language used in Artificial Intelligence."
text2 = "Python is widely used for AI and Machine Learning applications."

# Store text in a list
texts = [text1, text2]

# Convert text into TF-IDF vectors
tfidf = TfidfVectorizer()
vectors = tfidf.fit_transform(texts)

# Calculate cosine similarity
score = cosine_similarity(vectors[0], vectors[1])

# Display results
print("=" * 50)
print("        NLP (Cosine Similarity)")
print("=" * 50)

print("\nText 1:")
print(text1)

print("\nText 2:")
print(text2)

print("\nCosine Similarity Score:", round(score[0][0], 2))

# Interpretation
if score[0][0] >= 0.75:
    print("Interpretation: Both texts are highly similar.")
elif score[0][0] >= 0.40:
    print("Interpretation: Both texts are moderately similar.")
else:
    print("Interpretation: Both texts are weakly similar.")
```
